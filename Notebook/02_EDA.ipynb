```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Display plots inside the notebook
%matplotlib inline

# Better-looking plots
plt.style.use("ggplot")
sns.set_theme(style="whitegrid")
```

```python
df = pd.read_csv("../Dataset/dataset.csv")

df.drop(columns=["Unnamed: 0"], inplace=True)

df["date"] = pd.to_datetime(df["date"])

df["volatility"] = (df["high"] - df["low"]) / df["open"]
```

```python
print(df.shape)
```

```python
df.head()
```

```python
df.describe()
```

```python
plt.figure(figsize=(10,5))

sns.histplot(df["volatility"], bins=50)

plt.title("Distribution of Cryptocurrency Volatility")

plt.show()
```

```python
plt.figure(figsize=(10,4))

sns.boxplot(x=df["volatility"])

plt.title("Volatility Boxplot")

plt.show()
```

```python
plt.figure(figsize=(10,8))

sns.heatmap(df.corr(numeric_only=True),
            annot=True,
            cmap="coolwarm")

plt.title("Correlation Heatmap")

plt.show()
```

```python
plt.figure(figsize=(12,6))

df["crypto_name"].value_counts().head(10).plot(kind="bar")

plt.title("Top 10 Cryptocurrencies")

plt.xlabel("Cryptocurrency")

plt.ylabel("Count")

plt.show()
```

```python
btc = df[df["crypto_name"] == "Bitcoin"]

plt.figure(figsize=(14,6))

plt.plot(btc["date"], btc["close"])

plt.title("Bitcoin Closing Price")

plt.xlabel("Date")

plt.ylabel("Price")

plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.histplot(df["volume"], bins=40)

plt.title("Trading Volume Distribution")

plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.histplot(df["marketCap"], bins=40)

plt.title("Market Capitalization Distribution")

plt.show()
```
